# Nobel Prize 

The Nobel Prize has been among the most prestigious international awards since 1901. Each year, awards are bestowed in chemistry, literature, physics, physiology or medicine, economics, and peace. In addition to the honor, prestige, and substantial prize money, the recipient also gets a gold medal with an image of Alfred Nobel (1833 - 1896), who established the prize.

![](Nobel_Prize.png)

The Nobel Foundation has made a dataset available of all prize winners from the outset of the awards from 1901 to 2023. The dataset used in this project is from the Nobel Prize API and is available in the `nobel.csv` file in the `data` folder.

In this project, you'll get a chance to explore and answer several questions related to this prizewinning data. And we encourage you then to explore further questions that you're interested in!

## Step:1

> Import Dataset

```python
# Loading in required libraries
import pandas as pd
import seaborn as sns
import numpy as np

# Start coding here!
df = pd.read_csv("data/nobel.csv")
```

## Step:2

Which decade and Nobel Prize category combination had the highest proportion of female laureates?

```python
print("Dataset Info:")
print(df.info())
print("\nMissing Values:")
print(df.isnull().sum())
```

```python

# Question 1: Most commonly awarded gender and birth country
top_gender = df['sex'].value_counts().index[0]  # Most common gender
top_country = df['birth_country'].value_counts().index[0]  # Most common country

print(f"\n1. Most common gender: {top_gender}")
print(f"   Most common birth country: {top_country}")

# Question 2: Decade with highest ratio of US-born winners
df['decade'] = (df['year'] // 10) * 10

# calculate the ratio of US-born winners 
usa_decade_ratio = df[df['birth_country'] == 'United States of America'].groupby('decade').size() / df.groupby('decade').size()
max_decade_usa = usa_decade_ratio.idxmax()

print(f"\n2. Decade with highest US-born winners ratio: {max_decade_usa}")

# Question 3: Decade and category with highest proportion of female laureates
# Female winners:
female_winners = df[df['sex'] == 'Female']

# Calculate the female proportion for each decade-category combination
female_ratio = female_winners.groupby(['decade', 'category']).size() / df.groupby(['decade', 'category']).size()

# Find the combination with the highest proportion.
max_female_combination = female_ratio.idxmax()
max_female_dict = {max_female_combination[0]: max_female_combination[1]}

print(f"\n3. Decade and category with highest female proportion: {max_female_dict}")

# Question 4: First woman to receive the Nobel Prize
# Sort the women winners by year
women_winners = df[df['sex'] == 'Female'].sort_values('year')
first_woman_name = women_winners.iloc[0]['full_name']
first_woman_category = women_winners.iloc[0]['category']

print(f"\n4. First woman Nobel winner: {first_woman_name}")
print(f"   Category: {first_woman_category}")

# Question 5: Individuals/Organizations with multiple Nobel Prizes
# Count the prizes for each winner
repeat_winners = df['full_name'].value_counts()
repeat_list = repeat_winners[repeat_winners > 1].index.tolist()

print(f"\n5. Multiple Nobel Prize winners: {repeat_list}")

# Store the final answers in variables
print("\n" + "="*50)
print("FINAL ANSWERS:")
print(f"top_gender = '{top_gender}'")
print(f"top_country = '{top_country}'")
print(f"max_decade_usa = {max_decade_usa}")
print(f"max_female_dict = {max_female_dict}")
print(f"first_woman_name = '{first_woman_name}'")
print(f"first_woman_category = '{first_woman_category}'")
print(f"repeat_list = {repeat_list}")
```
